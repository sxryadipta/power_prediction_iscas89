```
# Create project folder structure
import os
for folder in ['data/raw', 'data/processed', 'results/metrics',
               'results/shap', 'results/figures']:
    os.makedirs(folder, exist_ok=True)
print("✓ Folder structure created")
```

```
# Upload your train.csv and test.csv
from google.colab import files
print("Upload train.csv and test.csv")
uploaded = files.upload()

import shutil
for fname in uploaded:
    shutil.copy(fname, f'data/raw/{fname}')
    print(f"✓ Saved: data/raw/{fname}")
```

```
import pandas as pd
import numpy as np

train = pd.read_csv('data/raw/train.csv')
test  = pd.read_csv('data/raw/test.csv')

print("TRAIN shape:", train.shape)
print(train.head())
print()
print("TEST shape:", test.shape)
print(test.head())
```

```
# Dataset summary statistics
print("=== TRAIN SUMMARY ===")
print(train.describe().round(4))
print()
print("=== TEST SUMMARY ===")
print(test.describe().round(4))
```

```
# Check for missing values
print("Missing in train:", train.isnull().sum().sum())
print("Missing in test: ", test.isnull().sum().sum())
print()
print("Target (power) range in train:", train['power'].min(), "—", train['power'].max(), "W")
print("Target (power) range in test: ", test['power'].min(),  "—", test['power'].max(),  "W")
```

```
# Save processed data (combined for LOOCV later)
all_data = pd.concat([train, test], ignore_index=True)
all_data.to_csv('data/processed/all_circuits.csv', index=False)
train.to_csv('data/processed/train.csv', index=False)
test.to_csv('data/processed/test.csv',  index=False)
print("✓ Data saved to data/processed/")
```
