# 07 — Results Summary & Paper Tables
Consolidates all results into clean, paper-ready tables and downloads everything.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

split_df = pd.read_csv('results/metrics/split_metrics.csv')
loocv_df = pd.read_csv('results/metrics/loocv_metrics.csv')
shap_df  = pd.read_csv('results/shap/shap_summary.csv')

print("=== TABLE I: Model Performance (Train/Test Split) ===")
print(split_df.round(5).to_string(index=False))
print()
print("=== TABLE II: Model Performance (LOOCV) ===")
print(loocv_df.round(5).to_string(index=False))
print()
print("=== TABLE III: SHAP Feature Importance ===")
print(shap_df[['Feature_Label','SHAP_XGBoost','SHAP_RF',
               'Rank_XGBoost','Rank_RF']].to_string(index=False))
```

```
# ── Export LaTeX tables for direct paste into paper ──
print("=== LaTeX: Table I ===")
print(split_df.round(5).to_latex(index=False, caption='Model performance on test set',
      label='tab:split', float_format='%.5f'))
print()
print("=== LaTeX: Table II ===")
print(loocv_df.round(5).to_latex(index=False, caption='LOOCV model performance (n=25)',
      label='tab:loocv', float_format='%.5f'))
```

```
# ── Download all results as a zip ──
import shutil
from google.colab import files

shutil.make_archive('iscas89_results', 'zip', 'results')
files.download('iscas89_results.zip')
print("✓ Downloaded iscas89_results.zip")
```
